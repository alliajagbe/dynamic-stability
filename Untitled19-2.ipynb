```python
import gym
import numpy as np

def run_episode(env, policy):
    observation = env.reset()
    episode = []
    while True:
        action = policy(observation)
        next_observation, reward, done, _ = env.step(action)
        episode.append((observation, action, reward))
        observation = next_observation
        if done:
            break
    return episode

def simple_policy(observation):
    # Simple policy: move left if pole is leaning left, move right otherwise
    return 0 if observation[2] < 0 else 1

def state_to_tuple(state):
    return tuple(state.tolist())

def monte_carlo(env, policy, episodes):
    returns_sum = {}
    returns_count = {}
    value_function = {}
    for _ in range(episodes):
        episode = run_episode(env, policy)
        states, actions, rewards = zip(*episode)
        G = 0
        for t in range(len(episode) - 1, -1, -1):
            state = states[t]
            action = actions[t]
            reward = rewards[t]
            G += reward
            state_tuple = state_to_tuple(state)
            if (state_tuple, action) not in returns_sum:
                returns_sum[(state_tuple, action)] = 0
                returns_count[(state_tuple, action)] = 0
            returns_sum[(state_tuple, action)] += G
            returns_count[(state_tuple, action)] += 1
            value_function[state_tuple] = returns_sum[(state_tuple, action)] / returns_count[(state_tuple, action)]
    return value_function

if __name__ == "__main__":
    env = gym.make('CartPole-v1')
    optimal_value_function = monte_carlo(env, simple_policy, episodes=1000)
    print("Optimal Value Function:")
    print(optimal_value_function)
```

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import random, math

env = gym.make('CartPole-v1', render_mode='human')
bins = list(zip(env.observation_space.low, env.observation_space.high))
bins[1] = (-0.5, 0.5)
bins[3] = (-math.radians(50), math.radians(50))
n_bins = (1,1,6,3)
Q = np.zeros(n_bins + (env.action_space.n,))
alpha = 0.1
max_episodes = 1000
gamma = 0.99
max_time_steps = 250
epsilon = 0.1
n_streaks = 0
solved_time = 199
streak_to_end = 120




def discretize(state):
    indices = []
    for i in range(len(state)):
        if state[i] <= bins[i][0]:
            index = 0
        elif state[i] >= bins[i][1]:
            index = n_bins[i] - 1
        else:
            bound_width = bins[i][1] - bins[i][0]
            offset = (n_bins[i]-1) * bins[i][0] / bound_width
            scaling = (n_bins[i]-1) / bound_width
            index = int(round(scaling * state[i] - offset))
        indices.append(index)
    return tuple(indices)

def epsilon_greedy(Q, state, epsilon):
    if random.uniform(0, 1) < epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])


def train():
    for episode in range(max_episodes):
        observation = env.reset()
        # print("From the outer for loop",observation)
        start_state = discretize(observation[0])
        prev_state = start_state
        done = False
        time_step = 0

        while not done:
            env.render()
            action = epsilon_greedy(Q, prev_state, epsilon)
            observation, reward, done, info, _ = env.step(action)
            # print("From the inner while loop",observation)
            best_q = np.max(Q[discretize(observation)])
            Q[prev_state][action] += alpha * (reward + gamma * best_q - Q[prev_state][action])
            state = discretize(observation)
            prev_state = state
            time_step += 1

        if time_step >= solved_time:
            n_streaks += 1
        else:
            n_streaks = 0

        if n_streaks > streak_to_end:
            print(f'Solved in {episode} episodes')
            break

    env.close()

train()
```

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# Environment setup
env = gym.make('CartPole-v1')

def make_epsilon_greedy_policy(Q, epsilon, nA):
    """
    Creates an epsilon-greedy policy based on a given Q-function and epsilon.
    """
    def policy_fn(observation):
        A = np.ones(nA, dtype=float) * epsilon / nA
        best_action = np.argmax(Q[observation])
        A[best_action] += (1.0 - epsilon)
        return A
    return policy_fn

def generate_episode(policy):
    """
    Generates an episode using the provided policy.
    """
    episode = []
    state = env.reset()
    state = tuple(state)  # Convert state to tuple for hashability
    for t in range(100):
        action_probs = policy(state)
        action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
        next_state, reward, done, _ = env.step(action)
        next_state = tuple(next_state)  # Convert next_state to tuple for hashability
        episode.append((state, action, reward))
        if done:
            break
        state = next_state
    return episode

def monte_carlo_control(env, num_episodes, discount_factor=0.99, epsilon_start=1.0, epsilon_decay=0.995, epsilon_min=0.1, test_frequency=100):
    """
    Monte Carlo Control using Epsilon-Greedy policies.
    Returns the optimal policy and the Q-function.
    """
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    rewards_per_episode = []  # Store rewards per episode for visualization
    epsilon = epsilon_start

    for i_episode in range(1, num_episodes + 1):
        policy = make_epsilon_greedy_policy(Q, epsilon, env.action_space.n)

        episode = generate_episode(policy)
        rewards = sum(x[2] for x in episode)  # Total rewards for the episode
        rewards_per_episode.append(rewards)

        seen_state_action_pairs = set()
        for t, (state, action, reward) in enumerate(episode):
            sa_pair = (state, action)
            if sa_pair not in seen_state_action_pairs:
                # Sum of discounted returns
                G = sum(x[2]*(discount_factor**i) for i, x in enumerate(episode[t:]))
                returns_sum[sa_pair] += G
                returns_count[sa_pair] += 1.0
                Q[state][action] = returns_sum[sa_pair] / returns_count[sa_pair]
                seen_state_action_pairs.add(sa_pair)

        # Update epsilon for the epsilon-greedy policy
        epsilon = max(epsilon * epsilon_decay, epsilon_min)

        if i_episode % test_frequency == 0:
            mean_reward = test_policy(policy, env)
            print(f"Episode {i_episode}/{num_episodes}, Mean Reward: {mean_reward:.2f}")

    return make_epsilon_greedy_policy(Q, epsilon_min, env.action_space.n), Q, rewards_per_episode

# Test the policy
def test_policy(policy, env, n_episodes=100):
    """
    Tests the policy over a specified number of episodes and returns the average reward.
    """
    rewards = []
    for _ in range(n_episodes):
        state = env.reset()
        state = tuple(state)  # Convert state to tuple for hashability
        done = False
        total_reward = 0
        while not done:
            action_probs = policy(state)
            action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
            next_state, reward, done, _ = env.step(action)
            next_state = tuple(next_state)  # Convert next_state to tuple for hashability
            state = next_state
            total_reward += reward
        rewards.append(total_reward)
    return np.mean(rewards)

# Run Monte Carlo Control
policy, Q, rewards_per_episode = monte_carlo_control(env, num_episodes=5000)

# Visualization
plt.figure(figsize=(10,5))
plt.plot(rewards_per_episode, label='Rewards per episode')
plt.xlabel('Episodes')
plt.ylabel('Total Reward')
plt.title('Rewards Per Episode - Monte Carlo Simulation')
plt.legend()
plt.show()
```
